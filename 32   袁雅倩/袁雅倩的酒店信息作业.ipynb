```python
from selenium import webdriver
import re,time
driver=webdriver.Chrome()
url = 'https://map.baidu.com/search/%E9%85%92%E5%BA%97/@13395888.6,3526321.62,12.5z?querytype=s&c=179&wd=%E9%85%92%E5%BA%97&da_src=shareurl&on_gel=1&l=12&gr=1&b=(13369773.72546261,3509752.059097923;13416719.235732773,3534221.7266132585)&pl_data_type=hotel&pl_sub_type=%E9%85%92%E5%BA%97&pl_price_section=0,%2B&pl_sort_type=default&pl_sort_rule=0&pl_discount2_section=0,%2B&pl_groupon_section=0,%2B&pl_cater_book_pc_section=0,%2B&pl_hotel_book_pc_section=0,%2B&pl_ticket_book_flag_section=0,%2B&pl_movie_book_section=0,%2B&pl_business_type=hotel&pn=0&device_ratio=2'
driver.get(url)
```

```python
list = []
while True:
    html = driver.page_source
    reobj = re.compile(r'"list_page">([\d\D]*?)</a>[\d\D]*?<span>([\d\D]*?)<i class="seperate">[\d\D]*?</i>([\d\D]*?)</span>[\d\D]*?<span class="n-grey" title="([\d\D]*?)">')
    for match in reobj.finditer(html):
        a = match.group(1)
        b = match.group(2)
        c = match.group(3)
        d = match.group(4)
        C =(a,b,c,d)
        print(C)
        list.append(C)
    match = re.search('false;">下一页&gt;</a>', html)
    if match:
        driver.find_element_by_link_text(u"下一页>").click()
        time.sleep(2)
    else:
        break
```

```python
import codecs
f = codecs.open(r"C:\Intel\袁雅倩的酒店信息作业.csv","w","utf-8-sig")
f.write("酒店名字,评分,类型,地址\n")
for i in list:
   f.write(",".join(i)+"\n") 
f.close()
```
