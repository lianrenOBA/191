```python
import re,time
from selenium import webdriver
driver=webdriver.Chrome()
urls = [("https://pintia.cn/problem-sets/1111652100718116864/problems/type/6","函数题"),
        ("https://pintia.cn/problem-sets/1111652100718116864/problems/type/7","编程题")]
def get_html(url):
    driver.get(url)
    time.sleep(6)
    html=driver.page_source
    return html
def get_bh(html):
    list1=[]
    reobj = re.compile(r'<td class="answerIcon_1du7d"></td><td>([\d\D]*?)</td>')
    for match in reobj.finditer(html):
        bh=match.group(1)
        list1.append(bh)
    return list1
def get_lj(html):
    list2=[]
    list3=[]
    reobj = re.compile(r'<a href="/problem-sets/1111652100718116864/problems/([\d\D]*?)" ')
    for match in reobj.finditer(html):
        lj=match.group(1)
        list2.append(lj)
    for i in list2:
        t="https://pintia.cn"+i
        list3.append(t)
    return list3
def get_bt(html):
    list4=[]
    reobj = re.compile(r'<a href="/problem-sets/1111652100718116864/problems/1111653039147499521" class="">([\d\D]*?)</a>')
    for match in reobj.finditer(html):
        bt=match.group(1)
        list4.append(bt)
    return list4
def get_fs(html):
    list5=[]
    reobj = re.compile(r"</a></td><td>([\d\D]*?)</td>")
    for match in reobj.finditer(html):
        fs=match.group(1)
        list5.append(fs)
    return list5
def get_tgs(html):
    list6=[]
    reobj = re.compile(r"</a></td><td>[\d\D]*?<td>([\d\D]*?)</td>")
    for match in reobj.finditer(html):
        tgl=match.group(1)
        list6.append(tgl)
    return list6
def get_tjs(html):
    list7=[]
    reobj = re.compile(r"</a></td>[\d\D]*?</td>[\d\D]*?</td>([\d\D]*?)</td>")
    for match in reobj.finditer(html):
        tjs=match.group(1)
        list7.append(tjs)
    return list7
def get_tgl(html):
    list8=[]
    reobj = re.compile(r"</a></td>[\d\D]*?</td>[\d\D]*?</td>[\d\D]*?</td>([\d\D]*?)</td></tr>")
    for match in reobj.finditer(html):
        tgl=match.group(1)
        list8.append(tgl)
    return list8
result=[]
for url,category in urls:
    html=get_html(url)
    bh=get_bh(html)
    lj=get_lj(html)
    bt=get_bt(html)
    fs=get_fs(html)
    tgs=get_tgs(html)
    tjs=get_tjs(html)
    tgl=get_tgl(html)
    for a,b,c,d,e,f,g in zip(bh,lj,bt,fs,tgs,tjs,tgl):
        l=(category,a,b,c,d,e,f,g)
        result.append(l)
print("OK")
```
