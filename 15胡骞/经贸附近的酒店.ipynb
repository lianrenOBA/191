```python
from selenium import webdriver
driver=webdriver.Chrome()
url = "https://map.baidu.com/search/%E6%B5%99%E6%B1%9F%E7%BB%8F%E8%B4%B8%E8%81%8C%E4%B8%9A%E6%8A%80%E6%9C%AF%E5%AD%A6%E9%99%A2%E9%99%84%E8%BF%91%E7%9A%84%E9%85%92%E5%BA%97/@13401816.6,3523321.62,15z?querytype=s&da_src=shareurl&wd=%E6%B5%99%E6%B1%9F%E7%BB%8F%E8%B4%B8%E8%81%8C%E4%B8%9A%E6%8A%80%E6%9C%AF%E5%AD%A6%E9%99%A2%E9%99%84%E8%BF%91%E7%9A%84%E9%85%92%E5%BA%97&c=179&src=0&pn=0&sug=0&l=13&b=(13381511,3513774;13427591,3539022)&from=webmap&biz_forward=%7B%22scaler%22:1,%22styles%22:%22pl%22%7D&device_ratio=1"
driver.get(url)#进入网站
```

```python
import re
import time
html = driver.page_source
print("酒店名称",",评分:",",类型:",",地址:")
list=[]
for i in range(2):       #采取两页的酒店信息
    html = driver.page_source
    reobj = re.compile(r'data-detail-code="list_page">([\d\D]*?)</a>[\d\D]*?<span>([\d\D]*?)<i[\d\D]*?\|</i>([\d\D]*?)<[\d\D]*?title="([\d\D]*?)>')
    for match in reobj.finditer(html):
        # match start: match.start()
        # match end (exclusive): match.end()
        # matched text: match.group()
        a=match.group(1)
        b=match.group(2)
        c=match.group(3)
        d=match.group(4)
        t=(a,b,c,d)
        list.append(t)
    driver.find_element_by_link_text(u"下一页>").click()#点击下一页
    time.sleep(2)
    print(list)
```
