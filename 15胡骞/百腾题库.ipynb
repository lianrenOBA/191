```python
from selenium import webdriver
import re,time
driver = webdriver.Chrome()
urls = [("https://pintia.cn/problem-sets/1111652100718116864/problems/type/6","函数题"),
        ("https://pintia.cn/problem-sets/1111652100718116864/problems/type/7","编程题")]

def get_html(url):
    driver.get(url)
    time.sleep(5)
    html = driver.page_source
    return html

def extract_info(html,category):
    import re
    results=[]
    reobj = re.compile(r'<td class="answerIcon_1du7d"></td><td>([\d\D]*?)</td><td><a href="([\d\D]*?)" class="">([\d\D]*?)</a></td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td><td>([\d\D]*?)</td></tr>')
    for match in reobj.finditer(html):
        a=match.group(1)
        b=match.group(2)
        b="https://pintia.cn" + b
        c=match.group(3)
        d=match.group(4)
        e=match.group(5)
        f=match.group(6)
        g=match.group(7)
        print("标号:",a,"url:",b,"标题:",c,"分数:",d,"通过数:",e,"提交数:",f,"通过率:",g,category)
        results.append([category,a,b,c,d,e,f,g])
    return results

results = []
for t in urls:
    url , category = t
    html = get_html(url)
    li  = extract_info(html,category)
    for elem in li:
        results.append(elem)
print(results)
```
