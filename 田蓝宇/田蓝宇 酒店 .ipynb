```python
from selenium import webdriver
import re,time
driver=webdriver.Chrome()
url = 'https://map.baidu.com/search/%E9%85%92%E5%BA%97/@13401816.6,3523321.62,12z?querytype=nb&c=179&wd=%E9%85%92%E5%BA%97&da_src=shareurl&on_gel=1&l=12&gr=1&b=(13373592.6,3499609.62;13430040.6,3547033.62)&pl_data_type=hotel&pl_sub_type=%E9%85%92%E5%BA%97&pl_price_section=0,%2B&pl_sort_type=default&pl_sort_rule=0&pl_discount2_section=0,%2B&pl_groupon_section=0,%2B&pl_cater_book_pc_section=0,%2B&pl_hotel_book_pc_section=0,%2B&pl_ticket_book_flag_section=0,%2B&pl_movie_book_section=0,%2B&pl_business_type=hotel&uid=7a0cb8c0d67a634683a00c28&nb_x=13401816.6&nb_y=3523321.62&gr_radius=1000&pn=0&device_ratio=1'
driver.get(url)
```

```python
list = []
while True:
    html = driver.page_source
    reobj = re.compile(r'"list_page">([\d\D]*?)</a>[\d\D]*?</i>([\d\D]*?)</span>')
    for match in reobj.finditer(html):
        a = match.group(1)
        b = match.group(2)
        c = (a,b)
        print(a,b)
        list.append(c)
    match = re.search(r'<a href="javascript:void\(0\)" tid="toNextPage" class="next next-none">下一页&gt;</a>', html)
    if not match:
        driver.find_element_by_link_text(u"下一页>").click()
        time.sleep(2)
    else:
        break
```

```python
import codecs
f = codecs.open("D:\TLY.csv","w","utf-8-sig")
f.write("hotel1,address1,type1\n")
for i in list:
   f.write(",".join(i)+"\n") 
f.close()
```
